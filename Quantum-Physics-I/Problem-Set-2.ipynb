### Part A

```python
k = sympy.symbols('k', positive=True)
omega, v_p, v_g = sympy.symbols('omega, v_p, v_g', cls=sympy.Function)
omega, v_p, v_g = (var(k) for var in [omega, v_p, v_g])

h = sympy.symbols('h')
g = sympy.symbols('g')
```

#### Equations

```python
phase_velocity = (v_p, omega/k)
sympy.Eq(*phase_velocity)
```

```python
group_velocity = (v_g, omega.diff(k))
sympy.Eq(*group_velocity)
```

```python
dispertion = (omega**2, g * k * sympy.tanh(k*h))
sympy.Eq(*dispertion)
```

#### Solving

```python
sol = sympy.solve(
    map(lambda x: sympy.Eq(*x), [
        dispertion,
        phase_velocity
    ]), 
    (omega, v_p),
    dict=True
)
sol
```

```python
sympy.Eq(v_p, sol[-1][v_p])
```

```python
v_g_gravity = group_velocity[-1].subs(sol[-1]).doit().simplify()
sympy.Eq(group_velocity[0], v_g_gravity)
```

For the open ocean, $h$ will be very large.

```python
sympy.Eq(v_p, sympy.limit(phase_velocity[1].subs(sol[-1]), h, sympy.oo)).simplify()
```

```python
sympy.Eq(v_g, sympy.limit(v_g_gravity, h, sympy.oo)).simplify()
```

### Part A

We need to prove

$$ \langle \hat{p} \rangle = \int \hbar k | \tilde{\psi}(k) |^2 \mathrm{d}k   \tag{1} $$ 

We can start with our known integral equation for $\hat{p}$.

$$ \langle \hat{p}^n \rangle = \int \psi^{*}(x) \: (\hat{p})^n \: \psi(x) \, \mathrm{d}x   \tag{2} $$ 

We use the case $n=1$ for equation (2):

$$ \langle \hat{p} \rangle = \int \psi^{*}(x) \: (\hat{p}) \: \psi(x) \, \mathrm{d}x   \tag{3} $$ 

We are also given a definition for $\hat{p}$:

$$ \hat{p} = - i \hbar \frac{\partial}{\partial x}   \tag{4} $$

Subbing our given equation (4) into equation (3) gives us:

$$ \langle \hat{p} \rangle = - i \hbar \int_{-\infty}^{\infty} \psi^{*}(x) \frac{\partial}{\partial x} \psi(x) \, \mathrm{d}x   \tag{5} $$ 

We can now use the definition of the inverse Fourier transform:

$$ \psi(x) = \frac{1}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k   \tag{6} $$

Subbing our given equation (6) into equation (5) gives us:

$$ \langle \hat{p} \rangle = - \frac{i \hbar}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \psi^{*}(x) \frac{\partial}{\partial x} \int_{-\infty}^{\infty} e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k \, \mathrm{d}x   \tag{7} $$ 

And we can switch the order of linear differential operators (differentiation and integration) in equation (7):

$$ \langle \hat{p} \rangle = - \frac{i \hbar}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \psi^{*}(x) \int_{-\infty}^{\infty} \frac{\partial}{\partial x} e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k \, \mathrm{d}x   \tag{8} $$ 

And we can now apply the differential operator in equation (8):

$$ \langle \hat{p} \rangle = \frac{\hbar}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \psi^{*}(x) \int_{-\infty}^{\infty} k e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k \, \mathrm{d}x   \tag{9} $$ 

We can rearrange the order of integrals in equation (9):

$$ \langle \hat{p} \rangle = \frac{\hbar}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} \psi^{*}(x)  \, k e^{i k x} \tilde{\psi}(k) \, \mathrm{d}x \, \mathrm{d}k   \tag{10} $$ 

And factor out the terms independent of $k$ in equation (10):

$$ \langle \hat{p} \rangle = \frac{\hbar}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} k \, \tilde{\psi}(k) \int_{-\infty}^{\infty} e^{i k x} \, \psi^{*}(x)  \, \mathrm{d}x \, \mathrm{d}k   \tag{11} $$ 

This looks close to the Fourier transform in the remaining $x$ terms, we just need to pull out the complex conjugate operator from the $x$ terms in equation (11):

$$ \langle \hat{p} \rangle = \frac{\hbar}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} k \, \tilde{\psi}(k) \left( \int_{-\infty}^{\infty} e^{- i k x} \, \psi(x) \, \mathrm{d}x \right)^{*} \, \mathrm{d}k   \tag{12} $$ 

And now we have the form of the Fourier transform

$$ \tilde{\psi}(k) = \frac{1}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} e^{- i k x} \psi(x) \, \mathrm{d}x   \tag{13} $$

And so subbing equation (12) into equation (11) gives:

$$ \langle \hat{p} \rangle = \hbar \int_{-\infty}^{\infty} k \, \tilde{\psi}(k) \tilde{\psi}^{*}(k) \, \mathrm{d}k   \tag{14} $$ 

Which simplifies to the original equation we were searching for:

$$ \langle \hat{p} \rangle = \int \hbar k | \tilde{\psi}(k) |^2 \mathrm{d}k   \tag{1} $$

We can also do it manually:

Starting with the definition of the Dirac delta function: 

$$ \int_{-\infty}^{\infty} \delta(x) \, f(x) \, \mathrm{d}x = f(0) $$

And the condition given for expressions involving delta functions to be equal:

$$ \int_{-\infty}^{\infty} D_{1}(x) \, f(x) \, \mathrm{d}x = \int_{-\infty}^{\infty} D_{2}(x) \, f(x) \, \mathrm{d}x $$

For this problem we need to use:

$$  D_{1}(x) = x \delta (x) $$
$$  D_{2}(x) = 0 $$

We can sub our $D_{1}(x)$ and $D_{2}(x)$ definitions into our condition:

$$ \int_{-\infty}^{\infty} x \delta (x) \, f(x) \, \mathrm{d}x = \int_{-\infty}^{\infty} (0) \, f(x) \, \mathrm{d}x   \tag{4} $$

Simplifying with the $0$ on the right hand side:

$$ \int_{-\infty}^{\infty} x \delta (x) \, f(x) \, \mathrm{d}x = 0   \tag{5} $$

We can use the definition of the Dirac delta function, with $ x \, f(x) $:

$$ (0) \, f(0) = 0 $$

Which simplifies to:

$$ 0 = 0   $$

Q.E.D.

#### Part ii

$$  D_{1}(x) = \delta (x) $$
$$  D_{2}(x) = \delta (-x) $$

$$ \int_{-\infty}^{\infty} \delta (x) \, f(x) \, \mathrm{d}x = \int_{-\infty}^{\infty} \delta (-x) \, f(x) \, \mathrm{d}x $$

We can immediately use the delta function definition on the left hand side:

$$ f(0) = \int_{-\infty}^{\infty} \delta (-x) \, f(x) \, \mathrm{d}x $$

On the right hand side we need to use a $u$ substitution first:

$$ x = -u $$
$$ \mathrm{d}x = - \mathrm{d}u $$

$$ f(0) = f(0) $$

#### Part iii

$$  D_{2}(x) = \frac{1}{|c|} \delta (x) $$

$$ \int_{-\infty}^{\infty} \delta (c x) \, f(x) \, \mathrm{d}x = \frac{1}{|c|} \int_{-\infty}^{\infty} \delta (x) \, f(x) \, \mathrm{d}x $$

Apply the delta function definition to the right hand side:

We can use $ c \operatorname{sgn}(c) = |c| $ to simplify:

$$ \frac{1}{|c|} \int_{-\infty}^{\infty} \delta (u) \, f\left(\frac{u}{c}\right) \, \mathrm{d}u = \frac{1}{|c|} f(0) $$

And finish with applying the definition of the delta function once again:

$$ \frac{1}{|c|} f(0) = \frac{1}{|c|} f(0) $$

Q.E.D.

$$ x \in (-\infty, \infty) \to u \in (-\infty, \infty)$$

$$ \lim_{a \to 0} \frac{1}{\sqrt{\pi}} \int_{-\infty}^{\infty} e^{-u^2} f(a \, u) \, \mathrm{d}u = f(0) $$

$$ \frac{1}{\sqrt{\pi}} \int_{-\infty}^{\infty} e^{-u^2} f(0) \, \mathrm{d}u = f(0) $$

$$ \frac{1}{\sqrt{\pi}} f(0) \int_{-\infty}^{\infty} e^{-u^2} \, \mathrm{d}u = f(0) $$
