## Problem 1

### Part A

#### Vars

```python
k = sympy.symbols('k', positive=True)
omega, v_p, v_g = sympy.symbols('omega, v_p, v_g', cls=sympy.Function)
omega, v_p, v_g = (var(k) for var in [omega, v_p, v_g])

h = sympy.symbols('h')
g = sympy.symbols('g')
```

#### Equations

```python
phase_velocity = (v_p, omega/k)
sympy.Eq(*phase_velocity)
```

```python
group_velocity = (v_g, omega.diff(k))
sympy.Eq(*group_velocity)
```

```python
dispertion = (omega**2, g * k * sympy.tanh(k*h))
sympy.Eq(*dispertion)
```

#### Solving

```python
sol = sympy.solve(
    map(lambda x: sympy.Eq(*x), [
        dispertion,
        phase_velocity
    ]), 
    (omega, v_p),
    dict=True
)
sol
```

```python
sympy.Eq(v_p, sol[-1][v_p])
```

```python
v_g_gravity = group_velocity[-1].subs(sol[-1]).doit().simplify()
sympy.Eq(group_velocity[0], v_g_gravity)
```

For the open ocean, $h$ will be very large.

```python
sympy.Eq(v_p, sympy.limit(phase_velocity[1].subs(sol[-1]), h, sympy.oo)).simplify()
```

```python
sympy.Eq(v_g, sympy.limit(v_g_gravity, h, sympy.oo)).simplify()
```

### Part B

```python
sympy.Eq(v_p, sympy.limit(phase_velocity[1].subs(sol[-1]), h, 0))
```

```python
sympy.Eq(v_g, sympy.limit(v_g_gravity, h, 0))
```

Since the velocities are decreasing, the energy must go into wave amplitude.

## Problem 3

### Part A

```python
x = sympy.symbols('x')
psi = sympy.symbols('psi', cls=sympy.Function)
```

```python
normalization_condition = sympy.Eq(sympy.Integral(sympy.Abs(psi(x))**2, (x, -sympy.oo, sympy.oo)), 1)
normalization_condition
```

```python
f, u, a, b = sympy.symbols('f, u, a, b', cls=sympy.Wild)

integral_dimension = (sympy.Integral(f, (u, a, b)), f * u)
```

```python
from sympy.physics import units
```

```python
psi_dim = sympy.symbols('Psi', real=True)
psi_dim
```

```python
dims = {
    x: units.length,
    psi(x): psi_dim
}
```

```python
normalization_condition_dim = normalization_condition.replace(*integral_dimension).subs(dims)
normalization_condition_dim
```

```python
lhs_dims = units.systems.si.dimsys_SI.get_dimensional_dependencies(normalization_condition_dim.lhs)
rhs_dims = units.systems.si.dimsys_SI.get_dimensional_dependencies(normalization_condition_dim.rhs)
lhs_dims
```

```python
sol = sympy.solve(
    normalization_condition_dim, 
    (psi_dim,), 
    dict=True
)
sol
```

### Part A

We need to prove

$$ \langle \hat{p} \rangle = \int \hbar k | \tilde{\psi}(k) |^2 \mathrm{d}k   \tag{1} $$ 

We can start with our known integral equation for $\hat{p}$.

$$ \langle \hat{p}^n \rangle = \int \psi^{*}(x) \: (\hat{p})^n \: \psi(x) \, \mathrm{d}x   \tag{2} $$ 

We use the case $n=1$ for equation (2):

$$ \langle \hat{p} \rangle = \int \psi^{*}(x) \: (\hat{p}) \: \psi(x) \, \mathrm{d}x   \tag{3} $$ 

We are also given a definition for $\hat{p}$:

$$ \hat{p} = - i \hbar \frac{\partial}{\partial x}   \tag{4} $$

Subbing our given equation (4) into equation (3) gives us:

$$ \langle \hat{p} \rangle = - i \hbar \int_{-\infty}^{\infty} \psi^{*}(x) \frac{\partial}{\partial x} \psi(x) \, \mathrm{d}x   \tag{5} $$ 

We can now use the definition of the inverse Fourier transform:

$$ \psi(x) = \frac{1}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k   \tag{6} $$

Subbing our given equation (6) into equation (5) gives us:

$$ \langle \hat{p} \rangle = - \frac{i \hbar}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \psi^{*}(x) \frac{\partial}{\partial x} \int_{-\infty}^{\infty} e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k \, \mathrm{d}x   \tag{7} $$ 

And we can switch the order of linear differential operators (differentiation and integration) in equation (7):

$$ \langle \hat{p} \rangle = - \frac{i \hbar}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \psi^{*}(x) \int_{-\infty}^{\infty} \frac{\partial}{\partial x} e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k \, \mathrm{d}x   \tag{8} $$ 

And we can now apply the differential operator in equation (8):

$$ \langle \hat{p} \rangle = \frac{\hbar}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \psi^{*}(x) \int_{-\infty}^{\infty} k e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k \, \mathrm{d}x   \tag{9} $$ 

We can rearrange the order of integrals in equation (9):

$$ \langle \hat{p} \rangle = \frac{\hbar}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} \psi^{*}(x)  \, k e^{i k x} \tilde{\psi}(k) \, \mathrm{d}x \, \mathrm{d}k   \tag{10} $$ 

And factor out the terms independent of $k$ in equation (10):

$$ \langle \hat{p} \rangle = \frac{\hbar}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} k \, \tilde{\psi}(k) \int_{-\infty}^{\infty} e^{i k x} \, \psi^{*}(x)  \, \mathrm{d}x \, \mathrm{d}k   \tag{11} $$ 

This looks close to the Fourier transform in the remaining $x$ terms, we just need to pull out the complex conjugate operator from the $x$ terms in equation (11):

$$ \langle \hat{p} \rangle = \frac{\hbar}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} k \, \tilde{\psi}(k) \left( \int_{-\infty}^{\infty} e^{- i k x} \, \psi(x) \, \mathrm{d}x \right)^{*} \, \mathrm{d}k   \tag{12} $$ 

And now we have the form of the Fourier transform

$$ \tilde{\psi}(k) = \frac{1}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} e^{- i k x} \psi(x) \, \mathrm{d}x   \tag{13} $$

And so subbing equation (12) into equation (11) gives:

$$ \langle \hat{p} \rangle = \hbar \int_{-\infty}^{\infty} k \, \tilde{\psi}(k) \tilde{\psi}^{*}(k) \, \mathrm{d}k   \tag{14} $$ 

Which simplifies to the original equation we were searching for:

$$ \langle \hat{p} \rangle = \int \hbar k | \tilde{\psi}(k) |^2 \mathrm{d}k   \tag{1} $$ 

### Part B

This time we are looking to show:

$$ \langle \hat{p}^2 \rangle = \int \left( \hbar k  \right)^2 | \tilde{\psi}(k) |^2 \mathrm{d}k   \tag{15} $$ 

We use the case $n=2$ for equation (2):

$$ \langle \hat{p}^2 \rangle = \int \psi^{*}(x) \: (\hat{p})^2 \: \psi(x) \, \mathrm{d}x   \tag{16} $$

Subbing our given equation (4) into equation (16) gives us:

$$ \langle \hat{p}^2 \rangle = - \hbar^2 \int_{-\infty}^{\infty} \psi^{*}(x) \frac{\partial^2}{\partial x^2} \psi(x) \, \mathrm{d}x   \tag{17} $$ 

Subbing our given equation (6) into equation (17) gives us:

$$ \langle \hat{p} \rangle = - \frac{\hbar^2}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \psi^{*}(x) \frac{\partial^2}{\partial x^2} \int_{-\infty}^{\infty} e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k \, \mathrm{d}x   \tag{18} $$ 

And we can switch the order of linear differential operators (differentiation and integration) in equation (18):

$$ \langle \hat{p} \rangle = - \frac{\hbar^2}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \psi^{*}(x) \int_{-\infty}^{\infty} \frac{\partial^2}{\partial x^2} e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k \, \mathrm{d}x   \tag{19} $$ 

And we can now apply the differential operator in equation (19):

$$ \langle \hat{p} \rangle = \frac{\hbar^2}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \psi^{*}(x) \int_{-\infty}^{\infty} k^2 e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k \, \mathrm{d}x   \tag{20} $$ 

We can rearrange the order of integrals in equation (20):

$$ \langle \hat{p} \rangle = \frac{\hbar^2}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} \psi^{*}(x)  \, k^2 e^{i k x} \tilde{\psi}(k) \, \mathrm{d}x \, \mathrm{d}k   \tag{21} $$ 

And factor out the terms independent of $k$ in equation (21):

$$ \langle \hat{p} \rangle = \frac{\hbar^2}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} k^2 \, \tilde{\psi}(k) \int_{-\infty}^{\infty} e^{i k x} \, \psi^{*}(x)  \, \mathrm{d}x \, \mathrm{d}k   \tag{22} $$ 

This looks close to the Fourier transform in the remaining $x$ terms, we just need to pull out the complex conjugate operator from the $x$ terms in equation (22):

$$ \langle \hat{p} \rangle = \frac{\hbar^2}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} k^2 \, \tilde{\psi}(k) \left( \int_{-\infty}^{\infty} e^{- i k x} \, \psi(x) \, \mathrm{d}x \right)^{*} \, \mathrm{d}k   \tag{23} $$ 

And so subbing equation (12) into equation (23) gives:

$$ \langle \hat{p}^2 \rangle = \hbar^2 \int_{-\infty}^{\infty} k^2 \, \tilde{\psi}(k) \tilde{\psi}^{*}(k) \, \mathrm{d}k   \tag{24} $$ 

Which simplifies to the original equation we were searching for:

$$ \langle \hat{p}^2 \rangle = \int \left( \hbar k  \right)^2 | \tilde{\psi}(k) |^2 \mathrm{d}k   \tag{15} $$ 

### Part C

This time we are looking to show:

$$ \langle f( \hat{p} ) \rangle = \int f( \hbar k ) | \tilde{\psi}(k) |^2 \mathrm{d}k   \tag{25} $$ 

We can do this with the definition of a Taylor series centered around $0$

$$ f(x) = \sum_{j=0}^{\infty} \frac{f^{(j)}(0)}{j!} x^j   \tag{26} $$

We can sub equation (26) into the left hand side of equation (25):

$$ \langle f( \hat{p} ) \rangle = \langle \sum_{j=0}^{\infty} \frac{f^{(j)}(0)}{j!} (\hat{p})^j \rangle   \tag{27} $$

Since we are working with sums, we can distribute the expectation operator over the sum in equation (27):

$$ \langle f( \hat{p} ) \rangle = \sum_{j=0}^{\infty} \frac{f^{(j)}(0)}{j!} \langle (\hat{p})^j \rangle   \tag{28} $$

Subbing this definition into equation (2):

$$ \langle f( \hat{p} ) \rangle = \sum_{j=0}^{\infty} \frac{f^{(j)}(0)}{j!} \int \psi^{*}(x) \: (\hat{p})^j \: \psi(x) \, \mathrm{d}x   \tag{29} $$

Subbing our given equation (4) into equation (29) gives us:

$$ \langle f( \hat{p} ) \rangle = \sum_{j=0}^{\infty} \frac{(-i \hbar)^j f^{(j)}(0)}{j!} \int_{-\infty}^{\infty} \psi^{*}(x) \frac{\partial^j}{\partial x^j} \psi(x) \, \mathrm{d}x   \tag{30} $$ 

Subbing our given equation (6) into equation (30) gives us:

$$ \langle f( \hat{p} ) \rangle = \frac{1}{\sqrt{2 \pi}} \sum_{j=0}^{\infty} \frac{(-i \hbar)^j f^{(j)}(0)}{j!} \int_{-\infty}^{\infty} \psi^{*}(x) \frac{\partial^j}{\partial x^j} \int_{-\infty}^{\infty} e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k \, \mathrm{d}x   \tag{31} $$ 

And we can switch the order of linear differential operators (differentiation and integration) in equation (31):

$$ \langle f( \hat{p} ) \rangle = \frac{1}{\sqrt{2 \pi}} \sum_{j=0}^{\infty} \frac{(-i \hbar)^j f^{(j)}(0)}{j!} \int_{-\infty}^{\infty} \psi^{*}(x) \int_{-\infty}^{\infty} \frac{\partial^j}{\partial x^j} e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k \, \mathrm{d}x   \tag{32} $$ 

And we can now apply the differential operator in equation (32):

$$ \langle f( \hat{p} ) \rangle = \frac{1}{\sqrt{2 \pi}} \sum_{j=0}^{\infty} \frac{\hbar^j f^{(j)}(0)}{j!}  \int_{-\infty}^{\infty} \psi^{*}(x) \int_{-\infty}^{\infty} k^j e^{i k x} \tilde{\psi}(k) \, \mathrm{d}k \, \mathrm{d}x   \tag{33} $$ 

We can rearrange the order of integrals in equation (33):

$$ \langle f( \hat{p} ) \rangle = \frac{1}{\sqrt{2 \pi}} \sum_{j=0}^{\infty} \frac{\hbar^j f^{(j)}(0)}{j!} \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} \psi^{*}(x)  \, k^j e^{i k x} \tilde{\psi}(k) \, \mathrm{d}x \, \mathrm{d}k   \tag{34} $$ 

And factor out the terms independent of $k$ in equation (34):

$$ \langle f( \hat{p} ) \rangle = \frac{1}{\sqrt{2 \pi}} \sum_{j=0}^{\infty} \frac{\hbar^j f^{(j)}(0)}{j!} \int_{-\infty}^{\infty} k^j \, \tilde{\psi}(k) \int_{-\infty}^{\infty} e^{i k x} \, \psi^{*}(x)  \, \mathrm{d}x \, \mathrm{d}k   \tag{35} $$ 

This looks close to the Fourier transform in the remaining $x$ terms, we just need to pull out the complex conjugate operator from the $x$ terms in equation (35):

$$ \langle f( \hat{p} ) \rangle = \frac{1}{\sqrt{2 \pi}} \sum_{j=0}^{\infty} \frac{\hbar^j f^{(j)}(0)}{j!} \int_{-\infty}^{\infty} k^j \, \tilde{\psi}(k) \left( \int_{-\infty}^{\infty} e^{- i k x} \, \psi(x) \, \mathrm{d}x \right)^{*} \, \mathrm{d}k   \tag{36} $$ 

And so subbing equation (12) into equation (36) gives:

$$ \langle f( \hat{p} ) \rangle = \sum_{j=0}^{\infty} \frac{\hbar^j f^{(j)}(0)}{j!} \int_{-\infty}^{\infty} k^j \, \tilde{\psi}(k) \tilde{\psi}^{*}(k) \, \mathrm{d}k   \tag{37} $$ 

Which simplifies to the original equation we were searching for:

$$ \langle f( \hat{p} ) \rangle = \int f(\hbar k) | \tilde{\psi}(k) |^2 \mathrm{d}k   \tag{25} $$
